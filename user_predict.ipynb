```python
import pandas as pd

user = {'name': 'pippo', 'surname': 'pluto', 'gender': 'M', 'searching_gender': 'F', 'interests': 'Roller hockey, Movies, Gaming, Paragliding, Orienteering', 'age': '45', 'latitude': '42.4355', 'longitude': '-83.489'}

df_user = pd.DataFrame.from_dict(user, orient="index").T
clustered = pd.read_csv("clustered_dataset.csv")
df_user
```

```python
from sklearn.feature_extraction.text import CountVectorizer  

def split_interests(interests):
  tokens = interests.split(", ")
  interests_list = [] 
    
  for token in tokens:
      interests_list.append(token)
        
  return interests_list

vect = CountVectorizer(tokenizer=lambda interests: split_interests(interests), lowercase=True) 
vect_labels = vect.fit_transform(df_user["interests"]) 


df_user = pd.concat([df_user, pd.DataFrame(data=vect_labels.toarray(), columns=vect.get_feature_names_out())], axis=1)
df_user
```

```python
merged = pd.concat([clustered, df_user], ignore_index = True)
merged = merged.fillna(0)
merged
```

```python
del merged['name']
del merged['surname']
del merged['gender']
del merged['searching_gender']
del merged['age']
del merged['Cluster #']
del merged['interests']

merged
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() 
merged[merged.columns] = scaler.fit_transform(merged[merged.columns])

merged
```

```python
import pandas as pd

df = pd.read_csv("clustered_dataset.csv")
data = df.copy()
data
```

```python
del data['interests']
del data['name']
del data['surname']
del data['gender']
del data['searching_gender']
del data['age']
data
```

```python
from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler() 
labels = data.columns.drop('Cluster #')
labels
data[labels] = scaler.fit_transform(data[labels])

data
```

```python
X = data.drop(['Cluster #'], 1)
y = data['Cluster #']
```

# Split dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

# Build Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

#Make predition
y_pred = rf.predict(X_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

def report_classifier(classifier, y_test, y_pred):
    print(f"Accuracy score:  {accuracy_score(y_test, y_pred)}")
    print(f"Classification report\n {classification_report(y_test, y_pred)}")

    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix= cm, display_labels= classifier.classes_)
    disp.plot()
    plt.show()

report_classifier(rf, y_test, y_pred)
```

# Build KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knnc = KNeighborsClassifier()
knnc.fit(X_train, y_train)
y_pred = knnc.predict(X_test)

report_classifier(knnc, y_test, y_pred)
```

# Buil SVC Classifier

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

report_classifier(knnc, y_test, y_pred)
```

# Build MLP Classifier

```python
from sklearn.neural_network import MLPClassifier

mlpc = MLPClassifier()
mlpc.fit(X_train, y_train)
y_pred = mlpc.predict(X_test)

report_classifier(mlpc, y_test, y_pred)
```

# Build GradientBoosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = MLPClassifier()
gbc.fit(X_train, y_train)
y_pred = gbc.predict(X_test)

report_classifier(gbc, y_test, y_pred)
```

From these results, we can see that SVC model has the best accuracy, but it isn't enough. In fact, we trained all the models by using default parameters. We need to try new configurations to find out the best classifier for our problem. So we need to implement a grid search who specificies a different configuration of parameters. In specific, we want to focus on 3 models:
- Random Forest
- KNN
- SVC

# Grid Search 

## Random Forest

```python
from sklearn.model_selection import GridSearchCV

rf_parameters = {
    'n_estimators': [250, 500, 750, 1000, 1250],
    'max_features': ['sqrt', 'log2'],
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 10, None],
    'min_samples_leaf':[1,2,3],
    'min_samples_split':[1,2,3]
}

rfc_gs = GridSearchCV(estimator = rf, param_grid = rf_parameters, cv = 10)
rfc_gs.fit(X_train, y_train)
rfc_gs.best_params_
```

```python
rf = RandomForestClassifier(criterion='gini', max_features='log2', n_estimators=1000, min_samples_leaf=1, min_samples_split=1, max_depth=None)
rf.fit(X_train, y_train)

#Make predition
y_pred = rf.predict(X_test)
report_classifier(rf, y_test, y_pred)
```

From 0.44 to 0.49 of accuracy

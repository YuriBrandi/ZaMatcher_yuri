```python
import pandas as pd

df = pd.read_csv("clustered_dataset.csv")
data = df.copy()
data
```

```python
del data['interests']
del data['name']
del data['surname']
del data['gender']
del data['age']
del data['searching_gender']
data
```

```python
from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler() 
labels = data.columns.drop('Cluster #')
labels
data[labels] = scaler.fit_transform(data[labels])

data
```

```python
X = data.drop(['Cluster #'], 1)
y = data['Cluster #']
```

# Split dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Build Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)

#Make predition
y_pred = rf.predict(X_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

def report_classifier(classifier, y_test, y_pred):
    print(f"Accuracy score:  {accuracy_score(y_test, y_pred)}")
    print(f"Classification report\n {classification_report(y_test, y_pred)}")

    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix= cm, display_labels= classifier.classes_)
    disp.plot()
    plt.show()

report_classifier(rf, y_test, y_pred)
```

# Build KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knnc = KNeighborsClassifier()
knnc.fit(X_train, y_train)
y_pred = knnc.predict(X_test)

report_classifier(knnc, y_test, y_pred)
```

# Build SVC Classifier

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

report_classifier(svc, y_test, y_pred)
```

# Build MLP Classifier

```python
from sklearn.neural_network import MLPClassifier

mlpc = MLPClassifier()
mlpc.fit(X_train, y_train)
y_pred = mlpc.predict(X_test)

report_classifier(mlpc, y_test, y_pred)
```

# Build GradientBoosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()
gbc.fit(X_train, y_train)
y_pred = gbc.predict(X_test)

report_classifier(gbc, y_test, y_pred)
```

From these results, we can see that SVC model has the best accuracy, but it isn't enough. In fact, we trained all the models by using default parameters. We need to try new configurations to find out the best classifier for our problem. So we need to implement a grid search that from certain parameters given in input, it will return the best configuration for that classifier.

# Grid Search 

## Random Forest

```python
from sklearn.model_selection import GridSearchCV

rf_parameters = {
    'n_estimators': [250, 500, 750, 1000, 1250],
    'max_features': ['sqrt', 'log2'],
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 10, None]
}

rfc_gs = GridSearchCV(estimator = rf, param_grid = rf_parameters, cv = 10, verbose=2)
rfc_gs.fit(X_train, y_train)
rfc_gs.best_params_
```

```python
rf = RandomForestClassifier(criterion='gini', max_features='log2', n_estimators=500, max_depth=None)
rf.fit(X_train, y_train)

#Make predition
y_pred = rf.predict(X_test)
report_classifier(rf, y_test, y_pred)
```

From 0.50 to 0.52 of accuracy

## SVC

```python
from sklearn.model_selection import GridSearchCV

svc_params = {
    'C':[1, 10, 100, 1000],
    'gamma':[1, 0.1, 0.5, 0.08, 0.001, 0.01, 0.0001], 
    'kernel':['linear','rbf', 'poly', 'sigmoid'],
    'degree': [2, 3, 4],
}

scv_fs = GridSearchCV(estimator = svc, param_grid = svc_params, cv = 10, verbose=2)
scv_fs.fit(X_train, y_train)
scv_fs.best_params_
```

```python
svc = SVC(C=1, gamma=0.1, kernel='rbf', degree=2)
svc.fit(X_train, y_train)

#Make predition
y_pred = svc.predict(X_test)
report_classifier(svc, y_test, y_pred)
```

From ~0.6715 to ~0.6745 (~ +0.003)

## KNN

```python
from sklearn.model_selection import GridSearchCV

neighbors = range(1, 31)

knn_params = {
    'n_neighbors': list(neighbors)
}

knn_gs = GridSearchCV(estimator = knnc, param_grid = knn_params, cv = 10, verbose=2)
knn_gs.fit(X_train, y_train)
knn_gs.best_params_
```

```python
knnc = KNeighborsClassifier(n_neighbors=1)
knnc.fit(X_train, y_train)
y_pred = knnc.predict(X_test)

report_classifier(knnc, y_test, y_pred)
```

0,50 to 0,53 of accuracy

## MLP
